1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:

Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

```python
#создадим копию базы данных ддя преобразований
hh_df_mod=hh_df.copy()
```

```python


# создадим функцию для выборки степени образования из столбца "Образование и ВУЗ", 
# далее записи образуем по категориям (высшее", "неоконченное высшее", "среднее специальное" и "среднее")
def column_education (grade):
    # преобразуем стоку в список, к нижнему регистру, разобьём по словам и запишем первых два слова
    grade = grade.lower().split(' ')[:2]
    # соединим два слова из списка и вставим между слов пробел
    grade = grade[0]+ ' ' + grade[1]
    # выполним выборку по интересующим нас категориям
    if grade ==  'высшее образование':
        educ_grade = 'высшее'
    elif grade == 'неоконченное высшее':
        educ_grade = 'неоконченное высшее'
    elif grade == 'среднее специальное':
        educ_grade = 'среднее специальное'
    elif grade == 'среднее образование':
        educ_grade = 'среднее'
    return educ_grade
# создадим новый столбец с категорией образования
hh_df_mod['Образование'] =hh_df_mod['Образование и ВУЗ'].apply(column_education)

# Удаляем столбец 'Образование и ВУЗ'
hh_df_mod = hh_df_mod.drop('Образование и ВУЗ', axis = 1)
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
# создадим функцию для выборки признака 'Пол'
def gender (gender_text):
    # разобьём  строку по словам, разделитель запятая. Из полученного списка сохраним первый элемент в переменную пол
    gender = gender_text.split(',')[0].strip(' ')
    if gender == 'Женщина':
        result = 'Ж'
    if gender == 'Мужчина':
        result = 'М'
    return result

# создадим функцию для выборки признака 'Возраст'
def age (age_text):
    # разобьем строку по словам, разделитель запятая. Из полученного списка сохраним второй элемент в переменную возраст
    age = age_text.split(',')[1]
    age = int(list(age)[2]+list(age)[3])
    return age

# применим функции к столбцу Пол, возраст. Сохраним данные в новые столбцы пол и возраст
hh_df_mod['Пол'] = hh_df_mod['Пол, возраст'].apply(gender)
hh_df_mod['Возраст'] = hh_df_mod['Пол, возраст'].apply(age)

# Удаляем столбец 'Пол, возраст'
hh_df_mod = hh_df_mod.drop('Пол, возраст', axis = 1) 
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# создадим функцию, которая позволит из столбца "опыт работы" получить переменную опыт работы в месяцах
def experience(exp_text): 
    if exp_text == 'Не указано' or exp_text is np.nan:
        result = np.nan
    else: 
        # для тех случаев года не указан месяц либо год, зададим переменные равные нулю
        mouth = 0
        year = 0
        exp = exp_text 
        exp = exp.split(' ')[2:6] # создаем список из строки, разделитель пробел. Нас интересует только часть текста с годом и месяцем 
        if exp [1][:3] == 'мес': 
            if exp[0].isalnum:
                mouth = exp[0]
        elif exp[1][:3] == 'год' or exp[1][:3] == 'лет':
            if exp[0].isalnum:
                year = exp[0]
            if exp [3][:3] == 'мес':
                if exp[0].isalnum:
                    mouth = exp[2]
        result = int(year)*12 + int(mouth)
    return result
# применим функцию к столбцу "опыт работы", результат запишем в столбец "Опыт"
hh_df_mod['Опыт'] = hh_df_mod['Опыт работы'].apply(experience)

# Удаляем столбец 'Опыт работы'
hh_df_mod = hh_df_mod.drop('Опыт работы', axis = 1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Создадим функцию которая будет возвращать True если указный тип занятости соответствует названию столбца
def one_hot_encoding(arg, key_column):
    result = False
    arg = arg.split(',') # создадим список разбив текст на отдельные фразы, разделитель запятая
    for i in range(len(arg)):
        arg[i] = arg[i].strip(' ')# избавимся от пробелов в начале и в конце строки
        if arg[i] == key_column: 
            result = True
            break
        else: result = False
    return result
```

```python
# Создадим список уникальных значений типов занятости
employment_info =  hh_df_mod['Занятость'].unique()
# создадим пустой лист и в него добавим тип занятости
employment_list = []
for text in employment_info:
        text = text.split(',')
        for word in text:
                word = word.strip(' ')
                if word not in employment_list:
                        employment_list.append(word)


# создадим колонки в соответствии полученного списка возможных вариантов типов занятости 
        # и применим к каждому созданному столбцу функцию one_hot_encoding
for column_emp in employment_list:
    hh_df_mod[column_emp] = hh_df_mod['Занятость'].apply(lambda x: one_hot_encoding(x,column_emp))
```

```python
# Создадим список уникальных значений типов "графика работы"
schedule_info =  hh_df_mod['График'].unique()
# создадим пустой лист и в него добавим тип "графика работы"
schedule_list = []
for text in schedule_info:
        text = text.split(',')
        for word in text:
                word = word.strip(' ')
                if word not in schedule_list:
                        schedule_list.append(word)
# создадим колонки в соответствии полученного списка возможных вариантов типов занятости 
        # и применим к каждому созданному столбцу функцию one_hot_encoding
for column_schedule in schedule_list:
    hh_df_mod[column_schedule] = hh_df_mod['График'].apply(lambda x: one_hot_encoding(x,column_schedule))
hh_df_mod.head()

# Удалим столбцы 'График' и 'Занятость'
hh_df_mod = hh_df_mod.drop(['График' , 'Занятость'], axis=1)
```

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и 
    # достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
hh_df_mod['Дата'] = pd.to_datetime(hh_df_mod['Обновление резюме']).dt.date
hh_df_mod['Дата'] = pd.to_datetime(hh_df_mod['Дата'] )
# Загружаем базу курса валют
currency = pd.read_csv('data/ExchangeRates.csv')
# Изменяем формат столбца в "DateTime "и сохраняем данные в новый столбец с названием таким же как в базе данных соискателей
currency ['Дата'] = pd.to_datetime(currency['date'])
# Создаем общий столбец для будущего объединения таблиц
currency['ЗП_валюта'] = currency['currency']
# удаляем не нужные столбцы
currency = currency.drop(['date', 'currency', 'vol', 'time', 'per'], axis = 1)

# создадим DataFrame, где индексы - вводимая соискателем валюта, столбец ISO - этаже валюта в стандарте ISO
currency_df = pd.DataFrame(
    index=currency_uniq, 
    data= ['RU','BYN','KZT','EUR','USD','UAH','UZS','KGS','AZN'],
    columns = ['ISO'])

#  Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется.
# Наименование валюты переведем в стандарт ISO согласно с таблицей выше.
salary_info = hh_df_mod['ЗП'].unique()
currency_uniq = []
for i in range(len(salary_info)):
    salary_info[i] = salary_info[i].split(' ')
    if salary_info[i][1] not in currency_uniq:
        currency_uniq.append(salary_info[i][1])

# Выделим из столбца "ЗП" данные по зарплате. Числовую величину сохраним в столбец "ЗП_сумма", а валюту в столбец "ЗП_валюта"  
hh_df_mod['ЗП_сумма'] = hh_df_mod['ЗП'].apply(lambda x:x.split(' ')[0])
hh_df_mod['ЗП_валюта'] = hh_df_mod['ЗП'].apply(lambda x: currency_df.loc[x.split(' ')[1]])


```

```python
# Объедим две таблицы, базу HH и курса валют. Общие столбцы: "Дата" и "ЗП_валюта" 
hh_df_mod = hh_df_mod.merge(currency, how= 'left', on = ['ЗП_валюта', 'Дата'])

# Значение close для рубля заполняем единицей 1 (курс рубля самого к себе)
hh_df_mod[['close', 'proportion']] = hh_df_mod[['close', 'proportion']].fillna(value=1)

# Умножим сумму желаемой заработной платы на присоединенный курс валюты (close) 
    # и разделим на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), 
        # результат занесем в новый столбец "ЗП (руб)"
# Поменяем тип данных "ЗП_сумма" в float
hh_df_mod['ЗП_сумма'] = hh_df_mod['ЗП_сумма'].astype('float')
hh_df_mod['ЗП (руб)'] = hh_df_mod['ЗП_сумма']*hh_df_mod['close']/hh_df_mod['proportion']
hh_df_mod['ЗП (руб)']
# Удалим столбецы  "ЗП", "Дата" ,"ЗП_сумма" , "ЗП_валюта ", "close ", "proportion"
hh_df_mod = hh_df_mod.drop([ 'ЗП', 'Дата' ,'ЗП_сумма' , 'ЗП_валюта', 'close', 'proportion'] , axis= 1)
```

На дух предоставленных графиках с нанесенными линиями границ 3-сигм, с послаблением в вправо на одну сигму.  Можно обратить внимание на следующее:
- Гистограмма распределения "Возраста" показала наличие выбросов данных начиная с 64 летнего возраста, при этом возраст соискателей моложе 18 лет находиться в пределах 3-ей сигмы(желтая штрих пунктирная линия), при уменьшении левой сигмы на единицу (зеленая штрих пунктирная линия), выбросы соискателей моложе 18 лет теперь за линией, что можно считать явными выбросами данных.
- На гистограмме распределения "Возраста" в логарифмическом масштабе, за пределами 3 сигмы слева, имеются выбросы молодых соискателей, при этом соискателе постарше находятся в пределах 4-ой сигмы справа. 


```python
def outliers_z_score_mod(data,feature, log_scale = False, left = 3,right = 3):
    if log_scale:
        x = np.log(data[feature])
    else: 
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left*sigma
    upper_bound = mu + right* sigma
    outler = data[(x<lower_bound)|(x>upper_bound)]
    cleaned = data[(x>lower_bound)&(x<upper_bound)]
    return outler , cleaned
outliers, cleaned = outliers_z_score_mod(hh_drop, 'Возраст', log_scale=True, right = 4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```
